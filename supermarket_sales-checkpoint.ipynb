```python
#renaming columns
ss_df.rename(columns={'Invoice ID': 'invoice_id', 'Customer type': 'customer_type', 'Product line': 'product_line', 'Unit price': 'unit_price', 'gross margin percentage': 'gross_margin_percentage', 'gross income': 'gross_income'}, inplace = True)
ss_df.head()
```

```python
#delete columns you don't need
ss_df=ss_df.drop(columns=['invoice_id','customer_type','Payment'])
ss_df.head(20)
```

```python
#Visualizing the correlation among all variables
sns.heatmap(np.round(ss_df.corr(),2), annot=True) 
```

```python
#to determine what product line sold the most
plt.figure(figsize=(20,10))
sns.histplot(data=ss_df, x="product_line", kde=True, color="blue")
```

```python
#to determine which branch made the most sales
plt.figure(figsize=(15,10))
sns.countplot(x=ss_df['Branch'])
```

```python
#to determine which branch made the highest income 
plt.figure(figsize=(15,10))
sns.boxplot(x=ss_df['Branch'], y=ss_df['gross_income'])
```

```python
#to determine which product sold the highest in the most profitable branch
plt.figure(dpi=125)
sns.countplot(y ='product_line', hue = "Branch", data = ss_df) 
plt.xlabel('Count')
plt.show()
```
